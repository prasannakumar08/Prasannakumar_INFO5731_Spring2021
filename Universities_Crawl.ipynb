```python
##Question 2 (12 points): Write python code to collect 500 items of the data you plan to collect above.

import time
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
path = '/home/prasanna/crawl'
number = (list(range(0, 522, 18)))
a = []
b = []


for i in number:
    url = 'https://www.collegesimply.com/colleges/search?sort=&place=&gpa=&sat=&act=&field=computer-and-information-sciences-and-support-services&major=&radius=300&zip=&state=&size=&tuition-fees=&net-price=&start='+str(i)
    r = requests.get(url)
    soup = BeautifulSoup(r.content, "html.parser")
    
    for hit in soup.findAll("a", itemprop="url"):
        name_dict = hit.attrs
        onclick_str = name_dict['onclick']
        splt_name = onclick_str.split("'")
        name = splt_name[7]
        a.append(name)
        
    for ultag in soup.find_all('ul', {'class': 'list-inline text-sm mb-3 text-muted'}):
        for litag in ultag.find_all('li', {'class': 'list-inline-item mr-3'}):
            for data in litag.find_all('span', {'class': 'mr-1 h6 text-dark'}):
                score = (data.text)
                b.append(score)

univ_name = pd.DataFrame(a)
score = pd.DataFrame(np.array(b).reshape(-1,6),columns=['Enrollment','Net Range','Acceptance Rate','SAT Range','ACT Range','AVG GPA'])

univ = pd.concat([univ_name, score.reindex(univ_name.index)], axis=1)
univ= univ.rename(columns = {0:'University'})
```
