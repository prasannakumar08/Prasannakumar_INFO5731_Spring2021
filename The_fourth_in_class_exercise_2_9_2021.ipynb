## 1. Text Data Preprocessing
## 1.1 Basic feature extraction using text data

```python
import pandas as pd
import numpy as np

pd.options.display.max_rows=200

with open('testo.txt', 'r') as file:
    data = file.read()
    

df = pd.DataFrame(data.split('\n')).reset_index(drop=True).rename(columns={0: 'data'})

df = df[df['data']!='']
df = df[df['data']!='\xa0']
df = df[df['data']!='—']
print(' ')
print('1.  Text Data Preprocessing')
print('1.1 Basic feature extraction using text data')
print(' ')
#Number of sentences
sentence_count = len(df)
print('Number of sentences in the file - '+str(sentence_count))

#Number of words
df['totalwords'] = df['data'].str.count(' ') + 1
print('Number of words in the file - '+str(df['totalwords'].sum()))

#Number of characters
df['nameCount'] = df['data'].str.len()
print('Number of characters in the file - '+str(df['nameCount'].sum()))

#Average word length
avgCount = df["totalwords"].mean()
print('Average word length in the file - '+str((avgCount).round(2)))

#Number of stopwords
from nltk.corpus import stopwords    
stop_words = set(stopwords.words('english'))
df['stopWords'] = df['data'].apply(lambda x: len(set(x.split()) & stop_words))
print('Number of stopwords in the file - '+str(df['stopWords'].sum()))

#Number of special characters
import string
count = lambda l1,l2: sum([1 for x in l1 if x in l2])
df['specChar'] = df.data.apply(lambda s: count(s, string.punctuation))
print('Number of special characters in the file - '+str(df['specChar'].sum()))

#Number of numerics
df['count_numbers'] = df['data'].str.count('\d')
print('Number of numerics in the file - '+str(df['count_numbers'].sum()))

#Number of uppercase words
df['Uppercase'] = df['data'].str.findall(r'[A-Z]').str.len()
print('Number of uppercase words in the file - '+str(df['Uppercase'].sum()))

```

## 1.2 Basic Text Pre-processing of text data

```python
#lowercasing
res1 = df['data'].apply(lambda x: x.lower()).to_frame().rename(columns={"data":"lower_case"})
df = pd.concat([df,res1],axis=1)
df
```

```python
#stopwords removal
import nltk
from nltk.corpus import stopwords
#nltk.download('stopwords')
list_of_words = set(stopwords.words('english')) 
res2 =  df['lower_case'].apply(lambda x:' '.join([i for i in x.split() if i not in list_of_words])).to_frame().rename(columns={"lower_case":"stop_words"})
df = pd.concat([df,res2],axis=1)
df
```

```python
#frequent words top 10
frequent_words = list(pd.Series(' '.join(df['data']).split()).value_counts()[:10].index)
#rare words top 10
rare_words = list(pd.Series(' '.join(df['data']).split()).value_counts()[-10:].index)
```

```python
print(frequent_words)
```

```python
print(rare_words)
```

```python
#frequent words removal 
res3 = df['stop_words'].apply(lambda x: ' '.join([i for i in x.split() if i not in frequent_words])).to_frame().rename(columns={"stop_words":"frequent_words"})
df = pd.concat([df,res3],axis=1)
df
```

```python
#rare words removal
res4 = df['frequent_words'].apply(lambda x: ' '.join([i for i in x.split() if i not in rare_words])).to_frame().rename(columns={"frequent_words":"rare_words"})
df =pd.concat([df,res4],axis=1)
df
```

```python
#spelling correction
from textblob import TextBlob
res5 = df['rare_words'].apply(lambda x: str(TextBlob(x).correct())).to_frame().rename(columns={"rare_words":"spell_check"})
df = pd.concat([df,res5],axis=1)
df
```

```python
#tokenisation
from nltk import word_tokenize
res6 = df['spell_check'].apply(lambda x: word_tokenize(str(x))).to_frame().rename(columns={"spell_check":"tokenization"})
df = pd.concat([df,res6],axis=1)
df
```

```python
#stemming
from nltk.stem import PorterStemmer 
from nltk.tokenize import word_tokenize 
   
ps = PorterStemmer() 

# stemming
res7 = df['tokenization'].apply(lambda x: ' '.join([ps.stem(i) for i in x])).to_frame().rename(columns={"tokenization":"stemming"})
df = pd.concat([df,res7],axis=1)
df
```

```python
import textblob
res8 = df['tokenization'].apply(lambda x: ' '.join([textblob.Word(i).lemmatize() for i in x])).to_frame().rename(columns={"tokenization":"lemmitization"})
df = pd.concat([df,res8],axis=1)
df
```

## 2.1 Write a Python program to remove leading zeros from an IP address

```python
# Python program to remove leading zeros  
import re  
  
def removeZeros(ip): 
    new_ip = re.sub(r'\b0+(\d)', r'\1', ip)
    return new_ip  
      
ip ="100.010.003.400"  
print(removeZeros(ip)) 
  
ip ="001.200.101.007"
print(removeZeros(ip)) 
```
